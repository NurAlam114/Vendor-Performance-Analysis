```python
import pandas as pd
import os
from sqlalchemy import create_engine
import logging
import time

logging.basicConfig(
    filename='logs/ingestion_db.log',
    level=logging.DEBUG,
    format="%(asctime)s - %(levelname)s - %(message)s",
    filemode='a'
)

engine = create_engine('sqlite:///inventory.db') 

def ingest_db(df,table_name,engine):
    """
    Insert a DataFrame into the database table.
    """
    df.to_sql(table_name, con = engine, if_exists='replace', index=False, chunksize=10000)
    
    
def load_raw_data():
    """
    Read all CSV files from 'data' folder and load them into the database.
    """
    start = time.time()
    
    for file in os.listdir('data'):
        if file.endswith('.csv'):
            df = pd.read_csv('data/' + file)
            print(file)
            print(df.shape)
            logging.info(f'Ingesting {file} in db')
            ingest_db(df, file[:-4], engine)
    
    end = time.time()
    total_time = (end - start) / 60
    
    logging.info('Ingestion Complete')
    logging.info(f'Total time: {total_time:.2f} minutes')
    
if __name__ == '__main__':
    load_raw_data()
        
```
