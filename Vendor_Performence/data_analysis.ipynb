* How much % of the total spend each vendor is taking.

```python
q = """
WITH v AS (
  SELECT VendorNumber, VendorName,
         SUM(Dollars + COALESCE(Freight,0)) AS total_spend
  FROM vendor_invoice
  GROUP BY VendorNumber, VendorName
),
t AS (
  SELECT SUM(total_spend) AS grand_total FROM v
)
SELECT v.VendorNumber, v.VendorName, v.total_spend,
       ROUND(v.total_spend * 100.0 / t.grand_total, 2) AS spend_percentage
FROM v CROSS JOIN t
ORDER BY spend_percentage DESC;
"""
pd.read_sql(q, conn).head(10)
```

* unit cost by every vendor 

```python
q = """
SELECT
  VendorNumber,
  VendorName,
  SUM(Quantity) AS total_Quantity	,
  SUM(Dollars) AS product_dollars,
  SUM(COALESCE(Freight,0)) AS freight_dollars,
  ROUND( (SUM(Dollars) + SUM(COALESCE(Freight,0))) / NULLIF(SUM(Quantity),0), 4) AS unit_cost_incl_freight
FROM vendor_invoice
GROUP BY VendorNumber, VendorName
ORDER BY unit_cost_incl_freight DESC;
"""
pd.read_sql(q, conn).head(10)
```

* How much time delay to payment by evrey vendor

```python
q = """
SELECT
  VendorNumber,
  VendorName,
  ROUND(AVG(julianday(PayDate) - julianday(InvoiceDate)), 2) AS avg_payment_days,
  MIN(julianday(PayDate) - julianday(InvoiceDate)) AS min_days,
  MAX(julianday(PayDate) - julianday(InvoiceDate)) AS max_days,
  COUNT(*) AS invoices
FROM vendor_invoice
WHERE PayDate IS NOT NULL AND InvoiceDate IS NOT NULL
GROUP BY VendorNumber, VendorName
ORDER BY avg_payment_days DESC;
"""
pd.read_sql(q, conn).head(10)
```

* Standard price vs Actual price 
* which vendor is charging more then the standard price

```python
q = """
SELECT
  p.VendorNumber,
  p.VendorName,
  p.Brand,
  p.Description,
  p.Size,
  ROUND(AVG(p.PurchasePrice), 4) AS avg_actual_price,
  ROUND(AVG(pp.PurchasePrice), 4) AS avg_standard_price,
  ROUND(AVG(p.PurchasePrice - pp.PurchasePrice), 4) AS avg_price_variance
FROM purchases p
JOIN purchase_prices pp
  ON p.VendorNumber = pp.VendorNumber
 AND p.Brand = pp.Brand
 AND p.Size = pp.Size
GROUP BY p.VendorNumber, p.VendorName, p.Brand, p.Description, p.Size
ORDER BY avg_price_variance DESC;
"""
pd.read_sql(q, conn).head(10)
```

* Vendor-wise Profit Query

```python
q = """
SELECT
    p.VendorNumber,
    p.VendorName,
    ROUND(AVG(pp.Price), 2) AS avg_selling_price,
    ROUND(AVG(p.PurchasePrice), 2) AS avg_actual_cost,
    
    ROUND(AVG(pp.Price - p.PurchasePrice), 2) AS profit_per_unit,
    
    SUM(p.Quantity) AS total_Quantity,
    
    ROUND(SUM((pp.Price - p.PurchasePrice) * p.Quantity), 2) AS total_profit,
    
    ROUND(AVG((pp.Price - p.PurchasePrice) / pp.Price) * 100, 2) AS margin_percent

FROM purchases p
JOIN purchase_prices pp
  ON p.VendorNumber = pp.VendorNumber
 AND p.Brand = pp.Brand
 AND p.Size = pp.Size

GROUP BY p.VendorNumber, p.VendorName
ORDER BY total_profit DESC;
"""

pd.read_sql(q, conn).head(10)
```

* Check how much risk which vendor

```python
q = """
WITH profit AS (
  SELECT
    p.VendorNumber,
    ROUND(
      AVG((pp.Price - p.PurchasePrice) / NULLIF(pp.Price,0)) * 100,
      2
    ) AS margin_percent
  FROM purchases p
  JOIN purchase_prices pp
    ON p.VendorNumber = pp.VendorNumber
   AND p.Brand = pp.Brand
   AND (
        TRIM(p.Size) = TRIM(pp.Size)
        OR (
            LOWER(pp.Size) LIKE '%ml'
            AND TRIM(p.Size) =
              ROUND(CAST(REPLACE(LOWER(pp.Size),'ml','') AS REAL)/1000.0, 2) || 'L'
        )
   )
  WHERE pp.Price IS NOT NULL AND pp.Price > 0
  GROUP BY p.VendorNumber
),

base AS (
  SELECT
    VendorNumber,
    VendorName,
    SUM(Quantity) AS total_Quantity,
    SUM(Dollars + COALESCE(Freight,0)) AS total_spend,
    ROUND(
      (SUM(Dollars + COALESCE(Freight,0)) / NULLIF(SUM(Quantity),0)),
      4
    ) AS unit_cost_incl_freight,
    ROUND(
      AVG(julianday(PayDate) - julianday(InvoiceDate)),
      2
    ) AS avg_payment_days
  FROM vendor_invoice
  WHERE InvoiceDate IS NOT NULL
    AND PayDate IS NOT NULL
    AND julianday(PayDate) >= julianday(InvoiceDate)
  GROUP BY VendorNumber, VendorName
),

purchase_totals AS (
  SELECT
    VendorNumber,
    SUM(Quantity) AS TotalPurchaseQuantity,
    SUM(Dollars)  AS TotalPurchaseDollars,
    ROUND(AVG(PurchasePrice), 4) AS PurchasePrice
  FROM purchases
  GROUP BY VendorNumber
),

actual_price AS (
  SELECT
    VendorNumber,
    ROUND(AVG(Price), 4) AS ActualPrice,
    ROUND(AVG(Volume), 2) AS Volume
  FROM purchase_prices
  WHERE Price IS NOT NULL AND Price > 0
  GROUP BY VendorNumber
),

sales_totals AS (
  SELECT
    VendorNo AS VendorNumber,
    SUM(SalesQuantity) AS TotalSalesQuantity,
    SUM(SalesDollars)  AS TotalSalesDollars,
    ROUND(
      (SUM(SalesDollars) / NULLIF(SUM(SalesQuantity),0)),
      4
    ) AS TotalSalesPrice,
    ROUND(SUM(ExciseTax), 2) AS TotalExciseTax
  FROM sales
  GROUP BY VendorNo
),

freight_totals AS (
  SELECT
    VendorNumber,
    ROUND(SUM(COALESCE(Freight,0)), 2) AS FreightCost
  FROM vendor_invoice
  GROUP BY VendorNumber
),

ranked AS (
  SELECT
    b.*,
    p.margin_percent,
    PERCENT_RANK() OVER (ORDER BY total_spend DESC)            AS spend_risk,
    PERCENT_RANK() OVER (ORDER BY unit_cost_incl_freight DESC) AS unit_cost_risk,
    PERCENT_RANK() OVER (ORDER BY avg_payment_days DESC)       AS payment_risk
  FROM base b
  LEFT JOIN profit p
    ON b.VendorNumber = p.VendorNumber
)

SELECT
  r.VendorNumber,
  r.VendorName,

  pt.PurchasePrice,
  ap.ActualPrice,
  ap.Volume,
  pt.TotalPurchaseQuantity,
  pt.TotalPurchaseDollars,
  st.TotalSalesQuantity,
  st.TotalSalesDollars,
  st.TotalSalesPrice,
  st.TotalExciseTax,
  ft.FreightCost,

  r.total_Quantity,
  r.total_spend,
  r.unit_cost_incl_freight,
  r.avg_payment_days,
  r.margin_percent,
  ROUND((r.spend_risk + r.unit_cost_risk + r.payment_risk)/3.0, 3) AS risk_score,
  CASE
    WHEN ((r.spend_risk + r.unit_cost_risk + r.payment_risk)/3.0) >= 0.67 THEN 'HIGH'
    WHEN ((r.spend_risk + r.unit_cost_risk + r.payment_risk)/3.0) >= 0.34 THEN 'MEDIUM'
    ELSE 'LOW'
  END AS risk_band
FROM ranked r
LEFT JOIN purchase_totals pt
  ON r.VendorNumber = pt.VendorNumber
LEFT JOIN actual_price ap
  ON r.VendorNumber = ap.VendorNumber
LEFT JOIN sales_totals st
  ON r.VendorNumber = st.VendorNumber
LEFT JOIN freight_totals ft
  ON r.VendorNumber = ft.VendorNumber
ORDER BY risk_score DESC, r.total_spend DESC;
"""

vendor_scorecard = pd.read_sql(q, conn)
vendor_scorecard.head(10)
```

```python
vendor_scorecard['VendorName'] = vendor_scorecard['VendorName'].str.strip()
```

```python
# Gross Profit
vendor_scorecard["GrossProfit"] = (
    vendor_scorecard["TotalSalesDollars"] - 
    vendor_scorecard["TotalPurchaseDollars"]
)
```

```python
# Profit Margin (%)
vendor_scorecard["ProfitMargin"] = (
    vendor_scorecard["GrossProfit"] / 
    vendor_scorecard["TotalSalesDollars"]
) * 100
```

```python
vendor_scorecard["StockTurnover"] = (
   vendor_scorecard["TotalSalesQuantity"] / 
   vendor_scorecard["TotalPurchaseQuantity"]
)
```

```python
# Sales to Purchase Ratio
vendor_scorecard["SalestoPurchaseRatio"] = (
    vendor_scorecard["TotalSalesDollars"] / 
    vendor_scorecard["TotalPurchaseDollars"]
)
```
